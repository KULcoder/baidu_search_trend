```python
normalized_data = copy.deepcopy(data)
for column in data.columns:
    if column == 'Date':
        continue
    normalized_data[column] = normalized_data[column] / normalized_data[column].std()
```

```python
normalized_data
```

```python
log_data = copy.deepcopy(data)
for column in data.columns:
    if column == 'Date':
        continue
    log_data[column] = np.log(log_data[column])
log_data = log_data.replace(-np.inf, 0) # deal with negative infinity
```

```python
file_path = "data/china.json"
```

```python
with open(file_path, 'rb') as json_file:
    original_data = json.load(json_file)
```

```python
for item in original_data['features']:
    EN = map_CN_EN[item['properties']['name']]
    trend_raw = data[EN].tolist()
    trend_norm = normalized_data[EN].tolist()
    trend_log = log_data[EN].tolist()
    item['properties']['EN'] = EN
    item['fever'] = {}
    item['fever']['trend_raw'] = trend_raw
    item['fever']['trend_norm'] = trend_norm
    item['fever']['trend_log'] = trend_log
```

```python
save_path = "data/q1plot.json"
with open(save_path, 'w') as json_file:
    json.dump(original_data, json_file)
```

## Plot 3
- Target Data structure: Network
- Node Size: Average Search Index Amount (optional)
- Link Size: Coefficient 

```python
data = pd.read_csv('data/data_raw.csv')
data = data.set_index('Date')
```

```python
map_CN_EN = {
    "新疆维吾尔自治区": 'Xinjiang',
    "西藏自治区": 'Tibet',
    "内蒙古自治区": 'Inner Mongolia',
    "青海省": 'Qinghai',
    "四川省": 'Sichuan',
    "黑龙江省": 'Heilongjiang',
    "甘肃省": 'Gansu',
    "云南省": 'Yunnan',
    "广西壮族自治区": 'Guangxi',
    "湖南省": 'Hunan',
    "陕西省": 'Shaanxi',
    "广东省": 'Guangdong',
    "吉林省": 'Jilin',
    "河北省": 'Hebei',
    "湖北省": 'Hubei',
    "贵州省": 'Guizhou',
    "山东省": 'Shandong',
    "江西省": 'Jiangxi',
    "河南省": 'Henan',
    "辽宁省": 'Liaoning',
    "山西省": 'Shanxi',
    "安徽省": 'Anhui',
    "福建省": 'Fujian',
    "浙江省": 'Zhejiang',
    "江苏省": 'Jiangsu',
    "重庆市": 'Chongqing',
    "宁夏回族自治区": 'Ningxia',
    "海南省": 'Hainan',
    "北京市": 'Beijing',
    "天津市": 'Tianjin',
    "上海市": 'Shanghai'
}
```

```python
map_EN_CN = {v: k for k, v in map_CN_EN.items()}
```

```python
cov_matrix = data.cov() / data.cov().to_numpy().max()
```

```python
cov_matrix = data.corr() / data.corr().to_numpy().max()
```

```python
plot3_json = {}
```

```python
# deal with node values
nodes = []
id_count = 1
max_value = data.mean().to_numpy().max()
for province in data.mean().index:
    nodes.append({"id": id_count, "name": province, "CN_name": map_EN_CN[province], "size": data.mean()[province]/max_value})
    id_count += 1
```

```python
nodes
```

```python
cov_matrix
```

```python
map_province_num = {name: i+1 for (i, name) in enumerate(cov_matrix.index)}
```

```python
flag_matrix = np.ones((31, 31))
```

```python
# deal with linkes values, our network is undirected
links = []
for i in range(1 , len(cov_matrix) + 1):
    sorted_value = cov_matrix.iloc[i-1].sort_values(ascending=False)
    count = 0
    for item in sorted_value.index:
        if count >= 5: continue
        ident = map_province_num[item]
        if ident == i:
            continue
        if flag_matrix[i-1, ident-1] == 1:
            flag_matrix[i-1, ident-1] == 0
            value = sorted_value[item]
            count += 1
    
        links.append({
            "source": i,
            "target": ident,
            "value": value
        })
```

```python
links
```

```python
plot3_json['nodes'] = nodes
```

```python
plot3_json['links'] = links
```

```python
save_path = "data/q3plot.json"
with open(save_path, 'w') as json_file:
    json.dump(plot3_json, json_file)
```
